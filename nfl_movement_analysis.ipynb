# Importar Librerías

```r
library(dplyr)
library(ggplot2)
library(readr)
```

# Cargar Datos

```r
data <- read_csv("supplementary_data.csv")
head(data)
```

# Preprocesamiento de Datos

```r
data_clean <- data %>% drop_na(yards_gained)
data_clean <- data_clean %>% mutate(yards_scaled = scale(yards_gained))
```

# Análisis Exploratorio de Datos

```r
summary(data_clean$yards_gained)
ggplot(data_clean, aes(x = yards_gained)) + geom_histogram(bins = 30) + theme_minimal()
```

# Pruebas Estadísticas

```r
t.test(yards_gained ~ possession_team, data = data_clean %>% filter(possession_team %in% c("KC", "DET")))
anova_model <- aov(yards_gained ~ as.factor(quarter), data = data_clean)
summary(anova_model)
```

# Modelado Estadístico

```r
model <- lm(yards_gained ~ down + yards_to_go, data = data_clean)
summary(model)
```
